### 크롤링을 위한 라이브러리 import

```
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
```

### Q1. 미국 Customer Reports 사이트에서 아래 url로 제품 정보(html) 조회
* 'https://www.consumerreports.org/cro/a-to-z-index/products/index.htm'
* UserAgent 라이브러리를 이용해 chrome 에서 접속한 것처럼 header 정보 생성

```
ua = UserAgent()
header = {'user-agent':ua.chrome}
url = "http://www.consumerreports.org/cro/a-to-z-index/products/index.htm"
page = requests.get(url, headers = header)
```

### Q2. 제품의 이름 정보와 링크 정보 조회 하여 딕셔너리 객체(products)에 할당

```
# 서버의 response 정보를 이용해 response 객체 생성
response = requests.get(url)

# reponse 객체에서 html 코드에 해당 부분 추출
data = response.text

# bs4를 이용하여 html 코드를 soup 객체로 파싱
soup = BeautifulSoup(data, 'html.parser')

tags = soup.find_all('div', class_="crux-body-copy")
# print(tags)
products = {tag.a.string:tag.a['href'] for tag in tags}
```

### Q3. products 객체를 loop 돌면서 제품명 과 link 정보 출력 

```
for product_name, product_link in products.items():
  print(product_name, product_link)
```
